2. Read the Tom Hanks filmography CSV file.

```python

from pyspark.sql import SparkSession

# Create the Spark session
spark = SparkSession.builder \
    .appName("TomHanksFilmography") \
    .getOrCreate()
```

```python
# read csv
df = spark.read.csv("/Users/marcellaferreira/pyspark/database/database.csv", header=True, inferSchema=True)
```
